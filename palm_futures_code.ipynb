```python
import os
import pandas as pd
```

```python
# Step 1: Load all CSV files from the folder
folder_path = 'C:/Users/jizatt/Documents/Courses/FRE_501/futures_contracts/palm_forward'
files = os.listdir(folder_path)
csv_files = [file for file in files if file.endswith('.csv')]

```

```python
# Step 2: Initialize the "palm_futures" DataFrame with the "Time" column from "CUF19"
palm_futures = pd.read_csv(os.path.join(folder_path, 'CUF19.csv'), usecols=['Time'], engine = "python", skipfooter=1)

palm_futures
```

```python
# Step 3: Process and join data from other DataFrames
years = ['19', '20', '21', '22', '23']
months = ['f', 'g', 'h', 'j', 'k', 'm', 'n', 'q', 'u', 'v', 'x', 'z']

for year in years:
    for month in months:
        name = f'cu{month}{year}'
        file_name = f'{name}.csv'
        
        if file_name in csv_files:
            df = pd.read_csv(os.path.join(folder_path, file_name), usecols=['Time', 'Last'], engine = "python", skipfooter=1)
            df.rename(columns={'Last': name}, inplace=True)
            
            
            # Full Outer Join with "palm_futures" DataFrame
            palm_futures = pd.merge(palm_futures, df, on='Time', how='outer')

# Display the resulting DataFrame
print(palm_futures)
```

```python
# Step 4: Adjust datetime and sort

# Adjust "Time" column to actual datetime
palm_futures["Time"] = pd.to_datetime(palm_futures["Time"], format = "%m/%d/%Y")

# Sort columns by date
palm_futures = palm_futures.sort_values(by='Time')



palm_futures
```

```python
palm_futures.to_csv("palm_futures.csv")
```
