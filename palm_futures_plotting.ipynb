# Plotting

The plotting process will be to transpose each data series, then make a line plot of each. 

I built this as a function, so each year can be called directly. Because of the absence of the 15th in some months, the exact day to call varies. You can access it directly with the following:

`sept_2022.index[0][-2:]`

This gets the September 2022 startday, `'15'`

```python
# generic plot function

def plot_price_paths(march_dataset, sept_dataset, year, mar_day = 15, sept_day = 15, tick_interval = 5):
    # Set up plot
    plt.figure(figsize=(10, 6))
    plt.title(f"Forward Price Paths of March, September {year}")
    plt.xlabel("Date")
    plt.ylabel("Value")

    # Plot input data series
    plt.plot(march_dataset.T.index, march_dataset.T[f"{year}-03-{mar_day}"], label=f"March {year}")
    plt.plot(sept_dataset.T.index, sept_dataset.T[f"{year}-09-{sept_day}"], label=f"September {year}")

    # Show every fifth day to reduce x-axis clutter
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=tick_interval))

    # Display the plot
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 2019

plot_price_paths(march_dataset = march_2019, sept_dataset = sept_2019, year = 2019, mar_day = march_2019.index[0][-2:], sept_day = sept_2019.index[0][-2:])
```

### 6 Month Out Plots

I need to

1. Take each *March* and September price series
2. Remove the NaN values before that month. 

- March starts at index 2, September starts at index 8

3. Reduce it to the 18 observations after the start point

4. Subtract the first month value from all others as spot price

5. Plot both series, to see if future curve is same or different. 

- this plotting done from base, so they align in "number of months out" but not actual months


March horizontal index series: 2 + 12x

September horizontal index series: 8 + 12x

```python
def plot_price_paths_baseline(march_dataset, sept_dataset, year, mar_day = 15, sept_day = 15, tick_interval = 1):
    

    # index march and sept datasets horizontally to remove all values left of start points
    march_index = 2 + 12*(year - 2019)
    sept_index = 8 + 12*(year - 2019)

    march_dataset = march_dataset.iloc[:, march_index:march_index+12]
    sept_dataset = sept_dataset.iloc[:, sept_index:sept_index+12]

    # subtract basis price values
    march_basis = march_dataset.iloc[0,0]
    sept_basis = sept_dataset.iloc[0,0]

    # reduce march, sept datasets by basic price to get price changes
    march_dataset = march_dataset - march_basis
    sept_dataset = sept_dataset - sept_basis

   
    # Set up plot
    plt.figure(figsize=(10, 6))
    plt.title(f"Forward Price Paths of March, September {year}")
    plt.xlabel("Months Until Contract Expiration")
    plt.ylabel("Value Change")

    # Plot input data series
    plt.plot(range(march_dataset.shape[1]), march_dataset.T[f"{year}-03-{mar_day}"], label=f"March {year}")
    plt.plot(range(sept_dataset.shape[1]), sept_dataset.T[f"{year}-09-{sept_day}"], label=f"September {year}")

    # Show every fifth day to reduce x-axis clutter
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=tick_interval))

    # Display the plot
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plot_price_paths_baseline(march_dataset = march_2019, sept_dataset = sept_2019, year = 2019, mar_day = march_2019.index[0][-2:], sept_day = sept_2019.index[0][-2:], tick_interval = 1)
```
